```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning) 
```

```python
df = pd.read_csv('hotel_bookings.csv')
df.head()
```

### Data Understanding

```python
def data_understanding(df):
        print(f"{'*' * 50} Data Information {'*' * 68}")
        display(df.info())
        print(f"\n{'*' * 50} Checking For Duplicate Values {'*' * 55}")
        display(df.duplicated().sum())
        print(f"\n{'*' * 50} Description of Numerical Columns {'*' * 52}")
        display(df.describe())
        print(f"\n{'*' * 50} Description of Categorical Columns {'*' * 50}")
        display(df.describe(include = 'object'))
        print(f"\n{'*' * 50} DataFrame Summary {'*' * 50}")
        display(df.head())
    
data_understanding(df)
```

### Data Cleaning

```python
def cleaned_data(df):
    # Handling missing values
    df['country'].fillna('Unknown', inplace = True)
    df['agent'].fillna(df['agent'].mean().round(1), inplace = True)
    df['company'].fillna(df['company'].mean(), inplace = True)
    df['children'].fillna(df['children'].median(), inplace = True)

    # Drop Duplicates
    df.drop_duplicates(inplace = True)

    # date to Datetime
    df['reservation_status_date']= pd.to_datetime(df['reservation_status_date'])

    # Total People in Booking
    df['total_people'] = df['adults'] + df['children'] + df['babies']

    # Total Stay Duration
    df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

    return df
```

### Exploratory Data Analysis

```python
def visualizations(df):
    # What proportion of bookings were cancelled vs. not cancelled?
    plt.figure(figsize=(6,6))
    sns.countplot(data=df, x='is_canceled', palette='Set2')
    plt.title('Booking Status Distribution')
    plt.xlabel('Booking Status')
    plt.ylabel('Count')
    plt.show()

    # Which hotel type has higher cancellations?
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x='hotel', hue='is_canceled', palette='Set1')
    plt.title('Cancellations by Hotel Type')
    plt.xlabel('Hotel Type')
    plt.ylabel('Count')
    plt.legend(['Not Canceled', 'Canceled'])
    plt.show()

    # Do longer lead times increase the chance of cancellation?
    plt.figure(figsize=(8,6))
    sns.boxplot(data=df, x='is_canceled', y='lead_time', palette='Set3')
    plt.title('Lead Time vs. Cancellation')
    plt.xlabel('Booking Status')
    plt.ylabel('Lead Time (Days)')
    plt.show()

    # Are cancellations seasonal?
    monthly_data = df.groupby(['arrival_date_month', 'is_canceled']).size().unstack()

    # Sort months correctly
    months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
                    'August', 'September', 'October', 'November', 'December']
    monthly_data = monthly_data.reindex(months_order)
    
    monthly_data.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Paired')
    
    plt.title('Monthly Booking and Cancellation Trends')
    plt.xlabel('Month')
    plt.ylabel('Number of Bookings')
    plt.xticks(rotation=45)
    plt.show()

    # Where do most customers come from?
    top_countries = df['country'].value_counts().head(10)

    plt.figure(figsize=(10,6))
    sns.barplot(x=top_countries.index, y=top_countries.values, palette='coolwarm')
    
    plt.title('Top 10 Customer Countries')
    plt.xlabel('Country')
    plt.ylabel('Number of Bookings')
    plt.xticks(rotation=45)
    plt.show()

    # Which market segments have the highest cancellations?
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x='market_segment', hue='is_canceled', palette='viridis')
    
    plt.title('Market Segment vs. Cancellations')
    plt.xlabel('Market Segment')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(['Not Canceled', 'Canceled'])
    plt.show()

    # Which numerical features are most related to cancellations?
    plt.figure(figsize=(12,8))
    sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.2f')
    
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
visualizations(df)
```
